```python
import logging
from typing import Dict, Tuple, Union

import h5py

logging.basicConfig(level=logging.INFO, format="%(message)s")


def print_hdf5_variables(file_path: str) -> Dict[str, Union[Tuple[int, ...], str]]:
    """
    读取 HDF5 文件并输出所有变量（dataset）名称及其维度。

    Args:
        file_path: HDF5 文件路径。

    Returns:
        一个字典，键为变量名，值为变量维度（tuple）或类型说明。
    """
    variables = {}

    with h5py.File(file_path, "r") as f:
        def _visitor(name, obj):
            if isinstance(obj, h5py.Dataset):
                variables[name] = obj.shape
                logging.info(f"{name}: shape={obj.shape}")
            elif isinstance(obj, h5py.Group):
                variables[name] = "Group"

        f.visititems(_visitor)

    return variables


print_hdf5_variables("/data/zehao/20260125_Pr4310_superstructure/All_Data_testnew/4D region32 thick0p4 copy01/20260303_full_N22201_dp128_sparse32_p4_1obj_32slice_dz10_plr4e-4_oalr4e-4_oplr8e-4_slr8e-4_tlr8e-4_dzlr8e-4_dpblur1_orblur0.4_ozblur1_mamp0.3_2_oathr0.98_oposc_sng1.0_spr0.1_C10_140_aff1.00_0.00_588.0_0.0/model_iter0250.hdf5")
```
